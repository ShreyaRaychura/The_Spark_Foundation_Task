## Step 1- Importing necessary libraries

```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
#import sklearn.datasets as datasets 

from sklearn.model_selection import train_test_split
import sklearn.metrics as sm
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

## Step 2 - Loading Iris dataset

```python
# Loading the dataset
Iris = pd.read_csv("C:/Users/SANJAY R/Desktop/Iris.csv",index_col="Id")
Iris
```

```python
X= Iris.drop('Species', inplace=False, axis=1)
Y= Iris["Species"]
```

#### Y is defined as the target variable of Iris dataset

## Step 4 - Modelling the data

```python
# Spliting the data in training and testing dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.1, random_state= 1,stratify=Y)
```

```python
tree_clf= DecisionTreeClassifier(max_depth=5, random_state=42)
tree_clf.fit(X_train,Y_train)
Y_pred = tree_clf.predict(X_test)
```

```python
print("Training Complete")
```

## Step 5- Comparision between actual and predicted classification

```python
data = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
data
```

## Step 6- Checking Model Accuracy

```python
print("Model Accuracy:", sm.accuracy_score(Y_test, Y_pred))
```

#### The accuracy of model is 0.93 which means 93%. Hence, the model is acceptable.

```python
print("classification report",sm.classification_report(Y_test, Y_pred))
```

```python
print("Confusion Matrix\n", sm.confusion_matrix(Y_test, Y_pred))
```

## Step 7- Visualising the trained model

```python
Iris_feature= ["PetalWidth","PetalLength","SepalLength", "SepalWidth"]
print(Iris_feature)
```

```python
target_names= ["setosa","versicolor","virginica"]
```

```python
plt.figure(figsize=(14,12))
plot_tree(tree_clf,feature_names=Iris_feature, class_names=target_names, 
          filled=True,rounded= True);
```
